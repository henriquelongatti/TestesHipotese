# Ordenar por Renda  

```python
uf = df.sort_values(by='UF')
```

```python
v = uf['UF'].unique()
```

```python
type(v)
```

```python
plt.close('all')
fig, ax = plt.subplots(figsize=(16, 8))
plt.ylim(0, 5000)
plt.title("Distribuição de Renda da População Preta no Brasil", fontsize=20)
plt.xticks(rotation=90, fontsize=14)
plt.yticks(fontsize=14)
sns.boxplot(x='UF', y = 'Renda',hue='Sexo',order=v,ax= ax, data=df.query('Renda < 6000'))
```

## Criando um dataset para ordenar por renda mediana 

```python
df[df['UF'] == 'Acre']['Renda'].median()
```

```python
median = [];
for i in v:
   m =df[df['UF'] == i]['Renda'].median()
   median.append(m) 
```

```python
median
```

```python
renda_mediana = { 'UF': v,'Mediana': median}
renda_mediana
```

```python
df_median = pd.DataFrame(renda_mediana,index=None)
df_median
```

```python
df_median = df_median.sort_values(by='Mediana')
```

```python
plt.close('all')
fig, ax = plt.subplots(figsize=(16, 8))
plt.ylim(0, 5000)
plt.title("Distribuição de Renda da População Preta no Brasil", fontsize=20)
plt.xticks(rotation=90, fontsize=14)
plt.yticks(fontsize=14)
sns.boxplot(x='UF', y = 'Renda',hue='Sexo',order=df_median['UF'],ax= ax, data=df.query('Renda < 6000'))

fig.savefig('../dados/Imagens/boxplot_renda_pretos_H_M.png', bbox_inches='tight')
```

# Comparando Mulheres  Negras com Homens Brancos 

```python
df_p = pd.read_csv('../dados/dados_preta.csv')
df_b = pd.read_csv('../dados/dados_brancos.csv')
```

```python
df_p.head()
```

```python
df_b.head()
```

```python
df_p = df_p[df_p['Sexo'] == 'Feminino']
```

```python
df_b = df_b[df_b['Sexo'] == 'Masculino']
```

```python
df = pd.concat([df_p,df_b])
```

```python
plt.close('all')
fig, ax = plt.subplots(figsize=(16, 8))
plt.ylim(0, 5000)
plt.title("Distribuição de Renda da População Preta no Brasil", fontsize=20)
plt.xticks(rotation=90, fontsize=14)
plt.yticks(fontsize=14)
sns.boxplot(x='UF', y = 'Renda',hue='Sexo',order=df_median['UF'],ax= ax, data=df.query('Renda < 6000'))

```

```python
sexo = {'Feminino': 'Mulheres Pretas',
        'Masculino':'Homens Brancos'}
```

```python
df['Sexo'] = df['Sexo'].map(sexo)
```

```python
df.head()
```

```python
plt.close('all')
fig, ax = plt.subplots(figsize=(16, 8))
plt.ylim(0, 5000)
plt.title("Distribuição de Renda da População Preta no Brasil", fontsize=20)
plt.xticks(rotation=90, fontsize=14)
plt.yticks(fontsize=14)
sns.boxplot(x='UF', y = 'Renda',hue='Sexo',order=df_median['UF'],ax= ax, data=df.query('Renda < 6000'))
fig.savefig('../dados/Imagens/boxplot_renda_mulheres_Pretas_homens-brancos.png', bbox_inches='tight')
```
